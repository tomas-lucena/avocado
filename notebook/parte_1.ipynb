```python
path_read = path.join(path.input,'avocado.csv')

names = ['date','average_price','total_volume','PLU_4046','PLU_4225','PLU_4770','total_bags','small_bags','large_bags','XLarge_bags','type','year','region']

df = pd.read_csv(
        path_read,
        index_col=0,
        names=names,
        header=0
)

df = df.sort_values(
            by='date',
            ignore_index=True
)

df
```

## Analise de cada coluna

### average_price 

É uma coluna com dados discrétos, referente a média dos preços dos abacates vendidos

```python
_df = df.groupby('date')['average_price'].mean()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        y=_df,
        x=_df.index
    )
)

fig.show()
```

### total_volume

```python
_df = df.loc[df.region=='TotalUS',['date','total_volume','type']]

fig = px.bar(
        y=_df['total_volume'],
        x=_df['date'],
        color=_df.type
    )

fig.show()
```

```python
_df = df.loc[df.region!='TotalUS',['date','total_volume','region']]

fig = px.bar(
        y=_df['total_volume'],
        x=_df['date'],
        color=_df['region']
    )

fig.show()
```

```python
df_pivot = df.pivot(['date','type'],'region','total_volume')
df_pivot
```

```python
_df = df_pivot.groupby('date').sum()
_df
```

```python
_df_percentage = _df.apply(lambda x: x/x['TotalUS'],axis=1)
_df_percentage
```

```python
_df_percentage_sum = (_df_percentage>0.1).sum()

columns = _df_percentage_sum.loc[_df_percentage_sum>1].index
columns = list(_df_percentage.loc[:,columns].mean().sort_values().index)
columns.remove('TotalUS')
```

```python
fig = px.bar(_df.loc[:,columns])
fig.show()
```
