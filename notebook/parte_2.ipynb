```python
from os import path,getcwd
import pandas as pd
import numpy as np

from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio

pio.templates.default = "plotly_white"
```

Trazer todas as métricas abordadas hoje para uma amostra representativa do dataset, em torno de 30% da base, ou seja: 
- a média, 
- a mediana, 
- o boxplot dos principais dados categóricos, 
- a matriz de correlação da base,
- as estatísticas das distribuições

```python
CORES_PASTEL = ["rgb(102, 197, 204)", "rgb(246, 207, 113)", "rgb(254, 136, 177)", "rgb(135, 197, 95)", "rgb(158, 185, 243)"]
```

```python
path.root = path.join(getcwd(),'..')
path.data = path.join(path.root,'data')
path.output = path.join(path.data,'output')
path.input = path.join(path.data,'input')
```

```python
lista_estados = ['Chicago', 'HarrisburgScranton', 'Pittsburgh','Boise', 'LosAngeles', 'LasVegas', 'Atlanta','DallasFtWorth', 'MiamiFtLauderdale','StLouis','Syracuse', 'Sacramento', 'Boston', 'Charlotte','Spokane', 'Albany', 'Houston', 'SanFrancisco','RaleighGreensboro', 'BuffaloRochester','GrandRapids', 'Denver', 'SanDiego','BaltimoreWashington', 'Roanoke', 'RichmondNorfolk', 'Louisville', 'Seattle','CincinnatiDayton', 'NewYork', 'NewOrleansMobile', 'Indianapolis','Jacksonville', 'Columbus', 'Detroit', 'Philadelphia','PhoenixTucson', 'Nashville', 'Portland', 'HartfordSpringfield','Tampa', 'Orlando',]
lista_regioes = ['Southeast', 'Plains',  'Midsouth', 'NorthernNewEngland', 'Northeast', 'SouthCentral', 'GreatLakes', 'West','WestTexNewMexico',]
```

```python
path_read = path.join(path.input,'avocado.csv')

names = ['date','average_price','total_volume','PLU_4046','PLU_4225','PLU_4770','total_bags','small_bags','large_bags','XLarge_bags','type','year','region']

dtypes = {
    'total_volume':np.int64,
    'PLU_4046':np.int64,
    'PLU_4225':np.int64,
    'PLU_4770':np.int64,
    'total_bags':np.int64,
    'small_bags':np.int64,
    'large_bags':np.int64,
    'XLarge_bags':np.int64
}


df = pd.read_csv(
        path_read,
        index_col=0,
        names=names,
        header=0,
        parse_dates=['date'],
)

df = df.astype(dtypes)

df = df.sort_values(
            by='date',
            ignore_index=True
)

df
```

```python
mask_estados = (df.type == 'conventional') & df.region.isin(lista_estados)
mask_regiao = (df.type == 'conventional') & df.region.isin(lista_regioes)
mask_total = (df.type == 'conventional') & (df.region == 'TotalUS')




average_price_total   = df.loc[mask_total,['date','average_price']].groupby('date')['average_price'].mean()  
average_price_estados = df.loc[mask_regiao, ['date','average_price']].groupby('date')['average_price'].mean()
average_price_regioes = df.loc[mask_estados,['date','average_price']].groupby('date')['average_price'].mean()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=average_price_total.index,
        y=average_price_total,
        name="media de preço US",
        marker=dict(color=CORES_PASTEL[0])
    )
)

fig.add_trace(
    go.Scatter(
        x=average_price_regioes.index,
        y=average_price_regioes,
        name="media de preço regiões nacionais",
        marker=dict(color=CORES_PASTEL[1])
    )
)

fig.add_trace(
    go.Scatter(
        x=average_price_estados.index,
        y=average_price_estados,
        name="media de preço regiões metropolitanas",
        marker=dict(color=CORES_PASTEL[2])
    )
)

fig.show()
#px.line(
#    _df,
#    x='date',
#    y=['conventional','organic']#,'media']
#)
```

```python
mask_estados = (df.type == 'conventional') & df.region.isin(lista_estados)
mask_regiao = (df.type == 'conventional') & df.region.isin(lista_regioes)
mask_total = (df.type == 'conventional') & (df.region == 'TotalUS')




average_price_total   = df.loc[mask_total,['date','average_price']].groupby('date')['average_price'].mean()  
average_price_estados = df.loc[mask_regiao, ['date','average_price']].groupby('date')['average_price'].mean()
average_price_regioes = df.loc[mask_estados,['date','average_price']].groupby('date')['average_price'].mean()

average_price_regioes_mean  = average_price_regioes.rolling(13).mean().dropna()
average_price_regioes_std   = average_price_regioes.rolling(13).std().dropna()

fig = make_subplots(rows=2,cols=1)


fig.add_trace(
    go.Scatter(
        x=average_price_regioes.index,
        y=average_price_regioes,
        name="preço",
        marker=dict(color=CORES_PASTEL[0])
    )
)

fig.add_trace(
    go.Scatter(
        x=average_price_regioes_mean.index,
        y=average_price_regioes_mean,
        name="media do preço",
        marker=dict(color=CORES_PASTEL[1])
    )
)

fig.add_trace(
    go.Scatter(
        x=average_price_regioes_mean.index,
        y=average_price_regioes_mean+2*average_price_regioes_std,
        name="media do preço +2 desvios",
        marker=dict(color=CORES_PASTEL[2])
    )
)

fig.add_trace(
    go.Scatter(
        x=average_price_regioes_mean.index,
        y=average_price_regioes_mean-2*average_price_regioes_std,
        name="media do preço -2 desvios",
        marker=dict(color=CORES_PASTEL[3])
    )
)

fig.add_trace(
    go.Bar(
        x=average_price_regioes_mean.index,
        y=2*average_price_regioes_std,
        name="desvio padrão",
        marker=dict(color=CORES_PASTEL[4])
    ),row=2,col=1
)


fig.show()
```
